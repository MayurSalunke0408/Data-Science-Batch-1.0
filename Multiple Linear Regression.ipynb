### Simple Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('advertising (1).csv')
```

```python
df.head()
```

```python
df.shape
```

```python
#EDA
y = df['Sales']
x = df['TV']
```

```python
df.corr()
```

```python
plt.scatter(x,y)
plt.xlabel('TV advertisement Price')
plt.ylabel('Sales based on Advertising')
plt.title('Scatter plot between Sales and TV advertise')
```

```python
#training and testing process
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3, random_state=0)
```

```python
len(x_train)
```

```python
len(y_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
```

```python
x_train = np.array(x_train).reshape(-1,1)
```

```python
model.fit(x_train, y_train)
```

```python
model.coef_  #beta1
```

```python
model.intercept_  #beta0
```

```python
line = model.intercept_ + model.coef_* x
plt.scatter(df['TV'], df['Sales'], color = 'r')
plt.plot(x,line)
plt.xlabel('TV advertisement Price')
plt.ylabel('Sales based on Advertising')
plt.title('Scatter plot between Sales and TV advertise')
```
